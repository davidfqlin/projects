```python
X = train.drop(['target', 'ID_code'], axis = 1)
y = train.target
```

```python
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 827)
```

```python
X_train.head(10)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver = 'sag')
```

```python
logreg.fit(X_train, y_train)
```

```python
y_val = logreg.predict(X_test)
```

```python
from sklearn import metrics
matrix = metrics.confusion_matrix(y_test, y_val)
matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_val))
print("Precision:",metrics.precision_score(y_test, y_val))
print("Recall:",metrics.recall_score(y_test, y_val))
```

```python
y_val_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_val_proba)
auc = metrics.roc_auc_score(y_test, y_val_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
